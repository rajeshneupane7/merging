```python
from joblib import Parallel, delayed
import pandas as pd

def process_group(key, data1, df2):
    # Find the matching group in df2
    match = df2[
        (df2['HerdCode'] == key[0]) &
        (df2['AnimalIndex'] == key[1]) &
        (df2['LeftHerdSeq'] == key[2])
    ]
    if not match.empty:
        data1 = data1.copy()
        data1['Key'] = match['CAR1'].iloc[0]
        return data1
    return None

# Run parallel processing
results = Parallel(n_jobs=-1, backend="multiprocessing")(
    delayed(process_group)(key, data1, df2)
    for key, data1 in df.groupby(['HerdCode', 'AnimalIndex', 'LeftHerdSeq'])
)

# Filter out None and concatenate
temp = [r for r in results if r is not None]
final_df = pd.concat(temp, ignore_index=True)
```
