```python
import pandas as pd
from datetime import timedelta

# ==========================================
# 1. PRE-PROCESSING (DO THIS ONCE ONLY)
# ==========================================

# A. Convert IDs to Integers
# We coerce errors to NaN, fill NaNs with -1 (or 0), then convert to Integer
cols_to_fix = [
    (master_sheet, 'ID'), 
    (merged_df, 'animal_id'), 
    (combined_df, 'Cow'), 
    (one_month_before, 'Animal_ID')
]

for df, col in cols_to_fix:
    # Convert to numeric, turn errors to NaN, fill NaN with -1, make int
    df[col] = pd.to_numeric(df[col]).astype(int)

# B. Convert Dates to "Midnight" (00:00:00)
# .dt.normalize() removes the time component but keeps it as a valid datetime object
master_sheet['date'] = pd.to_datetime(master_sheet['date']).dt.normalize()
merged_df['date_x'] = pd.to_datetime(merged_df['date_x']).dt.normalize()
combined_df['date'] = pd.to_datetime(combined_df['date']).dt.normalize()
one_month_before['Date'] = pd.to_datetime(one_month_before['Date']).dt.normalize()

# ==========================================
# 2. THE LOOP
# ==========================================

temp_data = []

# Iterate through Master Sheet
for animal_id, data_animal in master_sheet.groupby('ID'):
    
    if data_animal.empty: continue
    
    # Calculate Window (using the normalized dates)
    # Note: removed the extra '))' from your original code
    last_date = data_animal.sort_values('date')['date'].iloc[-1]
    start_date = last_date - timedelta(days=60)
    end_date = last_date + timedelta(days=30)
    
    # Filter Subsets 
    # Since we fixed the dates at the top, we don't need to convert them here.
    
    subset1 = merged_df[
        (merged_df['animal_id'] == animal_id) & 
        (merged_df['date_x'] > start_date) & 
        (merged_df['date_x'] < end_date)
    ]
    
    subset2 = combined_df[
        (combined_df['Cow'] == animal_id) & 
        (combined_df['date'] > start_date) & 
        (combined_df['date'] < end_date)
    ]
    
    subset3 = one_month_before[
        (one_month_before['Animal_ID'] == animal_id) & 
        (one_month_before['Date'] > start_date) & 
        (one_month_before['Date'] < end_date)
    ]

    temp=subset1.merge(subset2, left_on=['animal_id', 'date_x'], right_on=['Cow', 'date'], how='outer')
    temp2= temp.merge(subset3, left_on=['animal_id', 'date_x'], right_on=['Animal_ID', 'Date'], how='outer')
    day0_rows = data_animal[data_animal['day'] == 0]

    if day0_rows.empty:
        #raise ValueError("ERROR: data_animal has NO rows where day == 0. Cannot compute day_0.")
        print('skipping animal')
        continue
    else:
        day_0 = day0_rows['date'].iloc[0]

        temp2['day'] = (temp2['date_x'] - day_0).dt.days

        
        cols = ['Bacteria', 'Gram', 'Type', 'baccure']
        reference = data_animal.set_index('day')[cols].to_dict('index')
        for c in cols:
            temp2['Bacteria']= None
        for i, row in temp2.iterrows():
            d= row['day']

            valid_keys=[k for k in reference.keys() if k<=d]
            if len(valid_keys) == 0:
            # If day < first checkpoint (day 0), use day 0
                ref_day = 0
            else:
                ref_day = max(valid_keys)
        
        # Fill the columns
            for c in cols:
                temp2.at[i, c] = reference[ref_day][c]
    temp_data.append(temp2)
```

```python
import pandas as pd
from datetime import timedelta

temp_data = []

for animal_id, data_animal in master_sheet.groupby('ID'):
    
    if data_animal.empty: continue
    
    # 1. Define Window
    last_date = data_animal.sort_values('date')['date'].iloc[-1]
    start_date = last_date - timedelta(days=30)
    end_date = last_date + timedelta(days=30)
    
    # 2. Filter Subsets
    subset1 = merged_df[
        (merged_df['animal_id'] == animal_id) & 
        (merged_df['date_x'] > start_date) & 
        (merged_df['date_x'] < end_date)
    ]
    
    subset2 = combined_df[
        (combined_df['Cow'] == animal_id) & 
        (combined_df['date'] > start_date) & 
        (combined_df['date'] < end_date)
    ]
    
    subset3 = one_month_before[
        (one_month_before['Animal_ID'] == animal_id) & 
        (one_month_before['Date'] > start_date) & 
        (one_month_before['Date'] < end_date)
    ]

    # 3. CHAINED OUTER MERGE WITH PATCHING
    
    # --- Merge 1 ---
    merged_result = data_animal.merge(
        subset1, 
        left_on=['ID', 'date'], 
        right_on=['animal_id', 'date_x'], 
        how='outer',
        suffixes=('', '_merged') 
    )
    # CRITICAL STEP: Fill the NaN 'ID' and 'date' so the next merge works
    merged_result['ID'] = merged_result['ID'].fillna(merged_result['animal_id'])
    merged_result['date'] = merged_result['date'].fillna(merged_result['date_x'])
    
    # --- Merge 2 ---
    merged_result = merged_result.merge(
        subset2, 
        left_on=['ID', 'date'], 
        right_on=['Cow', 'date'], 
        how='outer', 
        suffixes=('', '_combined')
    )
    # Patch again
    merged_result['ID'] = merged_result['ID'].fillna(merged_result['Cow'])
    # No need to patch date here if 'Cow' matches 'ID' implies the date is valid, 
    # but safest to patch date if subset2 introduced a new date:
    # Note: subset2 'date' column name clashes with master 'date'. 
    # Pandas likely renamed subset2's date to 'date_combined'. Check your column names.
    if 'date_combined' in merged_result.columns:
         merged_result['date'] = merged_result['date'].fillna(merged_result['date_combined'])

    # --- Merge 3 ---
    merged_result = merged_result.merge(
        subset3, 
        left_on=['ID', 'date'], 
        right_on=['Animal_ID', 'Date'], 
        how='outer',
        suffixes=('', '_month_prev')
    )
    # Patch one last time
    merged_result['ID'] = merged_result['ID'].fillna(merged_result['Animal_ID'])
    merged_result['date'] = merged_result['date'].fillna(merged_result['Date'])

    temp_data.append(merged_result)

# 4. Compile
if temp_data:
    final_df = pd.concat(temp_data, ignore_index=True)
    # Cleanup: Ensure ID is integer (Outer join converts int to float because of NaNs)
    final_df['ID'] = final_df['ID'].fillna(-1).astype(int)
    print("Outer merge complete.")
```
