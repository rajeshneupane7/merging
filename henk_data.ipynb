```python

import datetime 
import os
import pandas as pd
from datetime import datetime
temp_file=[]
directory='/home/rajesh/work/bhuwan'
afi_list = []


for files in os.listdir(directory):
    if files.startswith('TAMU'):
        file_path = os.path.join(directory, files)
        
        # Extract date from filename
        filename = os.path.basename(files)  # Get only the filename
        date_str = filename.split(" - ")[1].split(" ")[0]  # Extract '01-01-2024'

        # Convert to datetime object
        file_date = datetime.strptime(date_str, "%m-%d-%Y")

        # Read CSV and process
        temp_df = pd.read_csv(file_path)
        temp_df = temp_df.loc[2:].reset_index(drop=True)  # Drop first two rows
        temp_df["date"] = file_date  # Add date column

        afi_list.append(temp_df)
```

```python
combined_df= pd.concat(afi_list)
```
